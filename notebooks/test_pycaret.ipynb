### Pycaret pour déterminer les modèles à comparer

```python
from pycaret.classification import *
import pandas as pd

loan_df=pd.read_csv("../data/Loan_clean.csv")
# Initialisation
s = setup(data=loan_df, target='Loan_Status', session_id=42)

# Comparaison automatique de plusieurs modèles
best_model = compare_models()

# Évaluation détaillée du meilleur modèle
evaluate_model(best_model)
```
